```python
import requests
import re
from bs4 import BeautifulSoup
from urllib.parse import urlparse
```

```python
def get_domain(url):
    return urlparse(url).netloc


def get_links(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
    soup = BeautifulSoup(requests.get(url, headers=headers).text)
    return soup.find_all('link')


def get_domain_links(domain, links):
    return [a.get('href') for a in links if urlparse(a.get('href')).netloc == domain]
```

```python
links = get_domain_links('www.walmart.com.br', get_links('https://www.walmart.com.br/caixa-de-som-2-1-c3-tech-sp-10-bk-3w-rms-usb-micro-sd/5024538/pr'))
```
